```r
#Tree based model (decision tree)-- Random Forest ()
#100,200,300...1000..2000,3000
#install.packages("randomForest",repos="https://cran.rstudio.com/")_
library(randomForest)
```

```r
#Bagging-it is a sampling technique for each tree,sampling with repition
```

```r
numTrain = 10000
```

```r
rows = sample(1:nrow(train),numTrain)
```

```r
class(train$label)
```

```r
#factor variable
labels = as.factor(train[rows,1])
```

```r
train1 = train[rows,-1]
```

```r
#tuninng the model or improve accuracy of model:-
#ntrees, mtry
```

```r
rf = randomForest(train1,labels,xtest =test,ntree=100,keep.forest=FALSE) #mtry = # of columns
```

```r
predictions =data.frame(ImageId=1:nrow(test),Label=levels(labels)[rf$test$predicted])
```

```r
head(predictions)
```

```r
#If you identify importance of each of variable present in the data
imp_var = data.frame(importance(rf))
```

```r
varImpPlot(rf)
```

```r
#ntrees-- do iteratively build multiple models with different ntree values,
#mtry --classificaiton:- square root of # of variables present
mtry = tuneRF(train1,labels,ntreeTry=20,stepFactor=1.5,improve=0.01)
#OOB-- out of bag error-model is built on 2/3 of data while error is checked on 1/3 of data
```

```r
best_mtry = mtry[mtry[,2] == min(mtry[,2]),]
```

```r
print(best_mtry)
```
