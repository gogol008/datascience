```r
salary <- read.csv(file='/root/hackerday/06_san_francisco_salary/output/Salaries.csv', header=TRUE,sep=",")
```

```r
#factor -> numerical
salary$Benefits <- as.numeric(as.character(salary$Benefits)) 
salary$OtherPay <- as.numeric(as.character(salary$OtherPay)) 
salary$OvertimePay <- as.numeric(as.character(salary$OvertimePay)) 
salary$BasePay <- as.numeric(as.character(salary$BasePay)) 

salary$EmployeeName <- as.character(salary$EmployeeName) 
salary$JobTitle <- as.character(salary$JobTitle) 
```

```r
#replace null value with 0
salary$Benefits[is.na(salary$Benefits)] <- 0
salary$BasePay[is.na(salary$BasePay)] <- 0
salary$OvertimePay[is.na(salary$OvertimePay)] <- 0
salary$OtherPay[is.na(salary$OtherPay)] <- 0
```

```r
#install.packages("dplyr) :- install a package
library(dplyr)
library(ggplot2)
```

```r
ggplot(salary,aes(x=Year,y=BasePay)) + geom_boxplot(fill="#53cfff") + theme_light(base_size=16)+ylab("Base Salary in $")
```

```r
boxplot(salary$BasePay[salary$Year==2011],salary$BasePay[salary$Year==2012],salary$BasePay[salary$Year==2013],
        salary$BasePay[salary$Year==2014])
```

```r
#quantile plots to check normality of data
n=length(salary$BasePay)
probabilities = (1:n)/(n+1)
```

```r
normal.quantiles = qnorm(probabilities,mean(salary$BasePay,na.rm=T),sd(salary$BasePay,na.rm=T))
```

```r
plot(sort(normal.quantiles),sort(salary$BasePay),xlab="Theoretical Quantiles from Normal Distribution",
     ylab= "Sample Quantiles of BasePay",main="Normal Quantile-Quantile plot of Basepay")
```
