```r
## Decision Tree
library(rpart)
#fit<-rpart(target~feat_1+feat_2+feat_3+feat_4,data=train) # only for features -- sample 
#fit<-rpart(target~.,data=train[-1]) # only for features -- sample 
fit<-rpart(target~.,control = rpart.control(cp = 0.05),data=train[-1]) # only for features -- sample 
summary(fit)
```

```r
#summary(fit)
#?rpart
pred<-predict(fit,test)
#head(pred)
submission_id<-as.character(test$id)
decision_tree_sub<-data.frame(id=submission_id,pred)
head(decision_tree_sub)
write.csv(decision_tree_sub,file="decision_tree_sub.csv",row.names=F)
```
